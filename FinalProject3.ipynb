```python
# การเเสดง dataframe ผ่านตัวเเปร data ที่เราให้เเทนค่าเอาไว้ โดยให้ 
#data0 = ข้อมูลจำนวนประชากรในกรุ่งเทพมหานคร จำเเนกตามอายุเเละเขตปี 56 
#data1 = ข้อมูลจำนวนประชากรในกรุ่งเทพมหานคร จำเเนกตามอายุเเละเขตปี 57 
#data2 = ข้อมูลจำนวนประชากรในกรุ่งเทพมหานคร จำเเนกตามอายุเเละเขตปี 59
#data3 = ข้อมูลชุมชนเเออัด
#ถ้าค่าของ olderthan100 ของทุกปี มีค่ามากกว่า 12 เเปลว่ามีคนที่อายุมากกว่า100อยู่มาก ให้ค่า Over100 เป็น  yes ถ้าน้อยกว่า 12 ให้ค่า Over100 เป็น No
data0 = pd.read_csv(os.path.join(path,'poppulationbyage56dcode.csv'))
data1 = pd.read_csv(os.path.join(path,'poppulationbyage57dcode.csv'))
data2 = pd.read_csv(os.path.join(path,'poppulationbyage59dcode.csv'))
data3 = pd.read_csv(os.path.join(path,'ชุมชนแออัด.csv'))
```

เลือกข้อมูลที่ต้องการใช้

```python
data0_X = data0[['id','dcode','dcode_sex','othernationalities_56','sex','over100_56','olderthan100_56']]
data0_X
```

```python
data1_X = data1[['dcode_sex','othernationalities_57','over100_57','olderthan100_57']]
data1_X
```

```python
data2_X = data2[['dcode_sex','othernationalities_59','over100_59','olderthan100_59']]
data2_X
```

###รวมตาราง

```python
merged_table_dcode_sex = data0_X.merge(data1_X,how='left',left_on='dcode_sex',right_on='dcode_sex')
merged_table_dcode_sex  #ครั้งที่ 1 
```

```python
merged_table_dcode_sex2 = merged_table_dcode_sex.merge(data2_X,how='left',left_on='dcode_sex',right_on='dcode_sex')
merged_table_dcode_sex2  #ครั้งที่ 2
```

```python
all_X =merged_table_dcode_sex2.merge(data3,how='left',left_on='dcode',right_on='dcode')
all_X
```

# data minning

```python
A0 = all_X.drop(columns=['id','dcode_sex','เขต','over100_56','over100_57','over100_59','olderthan100_56','olderthan100_57','olderthan100_59'])
A0
```

```python
A0['morethan&nomorethan'] = A0['morethan&nomorethan'].map({'more': 1 ,'nomore': 2 })
A0 #ใช้คำสั่ง map เพื่อเปลี่ยนข้อมูลในคอลัม  morethan&nomorethan ให้เป็นตัวเลข โดยให้ more = 1 เเละ nomore = 2
```

```python
A1 = all_X.drop(columns=['id','dcode_sex','เขต','dcode','othernationalities_56','sex','othernationalities_57','othernationalities_59','slum','morethan&nomorethan','olderthan100_56','olderthan100_57','olderthan100_59'])
A1
```

```python
A1['over100_56'] = A1['over100_56'].map({'yes': 1 ,'no': 2 })
A1 
```

```python
A1['over100_57'] = A1['over100_57'].map({'yes': 1 ,'no': 2 })
A1 
```

```python
A1['over100_59'] = A1['over100_59'].map({'yes': 1 ,'no': 2 })
A1 
```

```python
y = A1.values
y
```

```python
X = A0.values
X
```

train_test_split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=6)
```

```python
X_train.shape
```

train - validation เเบ่งอีกรอบ

```python
X_train2, X_val, y_train2, y_val = train_test_split(X_train, y_train, test_size=0.5, random_state=6)
```

```python
X_train2.shape #เหลือเเค่ 25 เเถว 7 คอลัม ที่จะนำมาทำต้นไม้
```

import

```python
from sklearn.tree import DecisionTreeClassifier
```

Define

```python
mytree =  DecisionTreeClassifier(max_leaf_nodes=3,criterion='entropy')  #กำหนดให้มีมากสุด 5 node
```

Train

```python
mytree.fit(X_train2,y_train2) 
```

Evaluate วัดผล

```python
from sklearn.metrics import accuracy_score
```

```python
tree_result = mytree.predict(X_val) # เอาข้อมูลเรามา Test 
```

```python
accuracy_score(y_val, tree_result ) #ความเเม่นยำของต้นไม้ได้ 24 เปอร์เซ็น
```
