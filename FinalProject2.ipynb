##เลือกข้อมูลที่ต้องการใช้

```python
data0_RE = data0[['id','dcode','dcode_sex','othernationalities_56','sex']]
data0_RE
```

```python
data1_RE = data1[['dcode_sex','othernationalities_57']]
data1_RE
```

```python
data2_RE = data2[['dcode_sex','othernationalities_59']]
data2_RE
```

###รวมตาราง

```python
merged_table_dcode_sex = data0_RE.merge(data1_RE,how='left',left_on='dcode_sex',right_on='dcode_sex')
merged_table_dcode_sex  #ครั้งที่ 1 
```

```python
merged_table_dcode_sex2 = merged_table_dcode_sex.merge(data2_RE,how='left',left_on='dcode_sex',right_on='dcode_sex')
merged_table_dcode_sex2  #ครั้งที่ 2
```

รวมตารางครั้งที่ 3

```python
merged_table_dcode_sex3 =merged_table_dcode_sex2.merge(data3,how='left',left_on='dcode',right_on='dcode')
merged_table_dcode_sex3 # ให้ค่าสลัมถ้ามากกว่า 12 เเปลว่า ชุมชุนนั้นเป็นชุมชนเเออัด(More) ถ้าน้อยกว่า 12 จะไม่เป็นชุมชนเเออัด(Nomore)
```
